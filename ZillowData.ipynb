```python
z1 = zillow.copy()
```

```python
p=[]
for cols in z1.columns:
    p.append(str(cols))
    
z1.columns =p
```

```python
z1.columns
```

```python
#let us assume that 2017-06 is the current selling price for our dataset
z1['Current Selling Price'] = z1['2017-06-01 00:00:00']
z1.head()
```

```python
#average rate by which a property rises in that zip code
z1['Rate']=(z1['2017-06-01 00:00:00']-z1['2007-06-01 00:00:00'])/10

#impute the missing rates by its median value
z3 = z1[~z1['Rate'].isnull()]
import numpy as np
z1['Rate'].fillna(np.median(z3['Rate']),inplace=True)
```

```python
p = z1.groupby('RegionName')['Rate'].mean()
p=pd.DataFrame(p)
p.reset_index(inplace=True)
p.sort_values(by='Rate',ascending=False,inplace=True)
p1 = p[p['Rate']<0]
p1.shape
```
