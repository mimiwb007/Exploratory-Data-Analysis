```python
import pandas as pd

d1 = pd.read_csv(r"2BHK-NY-combined.csv")
```

```python
#find annual rent

def mak(col):
    p = str(col).replace('$','').replace(',','')
    return 0.75*365*float(p)


#find annual rent
d1['AnnualRentofProperty'] =d1['price'].apply(mak)
```

```python
#payback period
d1['Payback_Period']=d1['Current Selling Price']/d1['AnnualRentofProperty']
```

```python
#metric is return on investment in 10 years assuming rent appreciates at a constant rate without including the equity value of the property
def mak(col1,col2,col3):
    m = (1+float(col2)/100)
    p = float(col1)*(m**10)
    p1 = p/float(col3)
    return p1
    
d1['ROI']=d1.apply(lambda e: mak(e['AnnualRentofProperty'],e['Rate'],e['Current Selling Price']),axis=1)
```

```python
#return on investment through rental as well as equity appreciation i.e. the increase in the price of the house is also taken into account to calculate roi
def incInPrice(col1,col2):
     m = (1+float(col2)/100)
     p = float(col1)*(m**10)
     p1 = p - float(col1)
     return p1

d1['ROI_inc']=d1.apply(lambda e: incInPrice(e['Current Selling Price'],e['Rate'],),axis=1)

def comb(col1,col2):
    return (col1+col2)/2

d1['ROI_incc'] = d1.apply(lambda e: comb(e['ROI_inc'],e['ROI'],),axis=1)
d1.drop('ROI_inc',axis=1,inplace=True)
```

```python
#create a dataframe with the metrics
d3 = d1[['zipcode','Current Selling Price','Rate', 'AnnualRentofProperty', 'Payback_Period','ROI','ROI_incc']]
```

```python
p = d3.groupby('zipcode').mean()
p.to_csv(r"Zipcode-Metrics.csv")
```
